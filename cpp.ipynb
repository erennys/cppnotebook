## Değişken Oluştururken Dikkat Edilmesi Gerekenler:
1. **Anlamlı ve Açıklayıcı İsimler Kullan**: Değişken adı, değişkenin neyi temsil ettiğini açıkça belirtmeli (örneğin, `score` veya `userName`).
2. **Türkçe Karakterlerden Kaçın**: C++ değişken isimlerinde Türkçe karakterler kullanmaktan kaçının (örneğin, `ad` yerine `name`).
3. **Özel Karakterlerden Kaçın**: Değişken isimleri yalnızca harf, rakam ve alt çizgi (`_`) içerebilir. Özel semboller (`@`, `%` vb.) kullanılamaz.
4. **Rakamla Başlama**: Değişken ismi rakamla başlayamaz (örneğin, `1value` yerine `value1`).
5. **Anahtar Kelimelerden Kaçın**: `int`, `float`, `return` gibi dilin anahtar kelimeleri değişken adı olarak kullanılamaz.
6. **Alt Çizgi ile Başlama**: Alt çizgi ile başlayan isimler genellikle sistemde kullanılan özel değişkenler içindir. Normal değişkenler için önerilmez.
7. **Case Sensitivity**: C++ büyük-küçük harfe duyarlıdır, bu nedenle `VariableName` ile `variablename` farklıdır.
8. **Uzunluktan Kaçın**: Çok uzun değişken isimleri yazımı zorlaştırır. Kısa ve öz ama açıklayıcı isimler tercih edin.
9. **Konvansiyonlara Uygunluk**: Belirli bir kodlama standardı varsa (örneğin, camelCase veya snake_case), buna uyun.
10. **Global ve Yerel Değişkenlerde Tutarlılık**: Global değişken isimleri genelde daha genel isimlendirilir, yerel değişkenlerde ise anlamlı ve özgün isimler tercih edilir.

## Bitsel Operatörler

| Operatör | Adı                        | Örnek         | Açıklama                                                                      |
|----------|-----------------------------|---------------|-------------------------------------------------------------------------------|
| `&`      | Bit düzeyinde VE (AND)     | `a & b`       | `a` ve `b`'nin her bitini karşılaştırır; ikisi de 1 ise sonuç biti 1 olur     |
| `|`      | Bit düzeyinde VEYA (OR)    | `a | b`       | `a` ve `b`'nin her bitini karşılaştırır; en az biri 1 ise sonuç biti 1 olur   |
| `^`      | Bit düzeyinde Özel VEYA (XOR)| `a ^ b`     | Bitler farklıysa sonuç biti 1, aynıysa 0 olur                                |

````c++
#include <iostream>
using namespace std;

int not1 = 72;
int not2;
not2 = 51;

cout << "Notlar toplami: " << (not1 + not2) << endl;

int notOrtalamasi;
notOrtalamasi = (not1 + not2) / 2;

cout << "Notlar ortalamasi: " << notOrtalamasi << endl;

/* değişkenlerimiz int değerinde olduğu için tam sayı değere yuvarlayarak sonuç gösteriyor
eğer küsüratlı bir şekilde istersek notOrtalamasinin başına ve aritmetik işlemin başına float demeliyiz.*/

float newNotOrtalamasi = (float) (not1 + not2) / 2;

cout << "Yeni notlar ortalamasi: " << newNotOrtalamasi;
````

# Switch Case Yapısı
* İf ve else..if ikilisi, birçok alternatif arasında bir blok kodu yürütmenize izin verir.

Ancak if deyimi genellikle iç içe olduğundan Switch deyimi daha hızlıdır. Ayrıca, switch ifadesinin sözdizimi daha temiz ve anlaşılması kolaydır.

````c++
#include <iostream>

using namespace std;

int gun;
cout << "Haftanın gününü sayı olarak giriniz: ";
cin >> gun;

switch (gun)
{
    case 1:
        cout << "Pazartesi";
        break;
    case 2: 
        cout << "Salı";
        break;
    case 3:
        cout << "Çarşamba";
        break;
    case 4:
        cout << "Perşembe";
        break;
    case 5:
        cout << "Cuma";
        break;
    case 6: 
        cout << "Cumartesi";
        break;
    case 7:
        cout << "Pazar";
        break;
    default:
    cout << "Geçersiz sayı girdiniz! Bir hafta 7 gündür.";
}
````

# While Döngüsü
* While döngüsü test ifadesini değerlendirir.

Test ifadesi doğruysa (sıfır değil), döngü sırasındaki gövdenin içindeki kodlar çıkarılır. Test ifadesi tekrar değerlendirilir. İşlem, test ifadesi yanlış olana kadar devam eder.

Test ifadesi yanlış olduğunda, while döngüsü sonlandırılır.

```cpp
#include <iostream>

using namespace std;

int index = 0;

while (indwx <= 3)
{
    cout << "index değeri 3'den küçük" << endl;
}
cout << "programın sonu";

//Bu kod parçasında sonsuza kadar index değeri 3^den küçüktür yazacaktır. Hiçbir zaman programın sonu yazmayacaktır. Çünkü while içerisindeki ifade doğru  
//While ifadeseninn içerisinin doğru olmasından dolayı program sürekli kendini tekrar tekrar çalıştıracak.  
//Sonsuz loop

````c++
#include <iostream>

using namespace std;

int index = 0;

while (index <= 3)
{
    cout << "index değeri 3'den küçük." << endl;
    index++;
}

cout << "programın sonu";

//Programın sonsuz loopa girmesini önlemek adına index i sürekli 1 olarak atrırıyoruz while döngüsüne her girdiğinde index değerimiz 0 1 2 3 diye devam ediyor ne zamana kadar index <= 3 olana kadar devam eder.    
//Burada yaptığımız şey indexi sınırlandırdık.  
````

# Do While Döngüsü
* Do..while döngüsü, önemli bir fark ile while döngüsüne benzer.  
Do..while döngüsü test ifadesi kontrol edilmeden önce bir kez yürütülür. Dolayısıyla, do … while döngüsü en az bir kez yürütülür.    
Parantezlerin içindeki kod bloğu (döngü gövdesi) bir kez yürütülür.  
Ardından, test ifadesi değerlendirilir. Test ifadesi doğruysa, döngü gövdesi tekrar yürütülür. Bu işlem, test ifadesi 0 (yanlış) olarak değerlendirilinceye kadar devam eder.  
Test ifadesi yanlış (sıfır olmayan) olduğunda, do … while döngüsü sonlandırılır.  

````c++
#include <iostream>
using namespace std;

int sifre = 1234;
int input;

do {
    cout << "Lütfen sifreyi giriniz: ";
    cin >> input;
}
while (input != sifre); 

cout << "Hoşgeldiniz!";
    
````

# For Döngüsü
* Başlatma ifadesi yalnızca bir kez yürütülür.

Ardından, test ifadesi değerlendirilir. Test ifadesi yanlış (0) ise, döngü sonlandırılır. Ancak eğer test ifadesi doğruysa (sıfır değil), for döngüsünün gövdesindeki kodlar yürütülür ve güncelleme ifadesi güncellenir.

Bu işlem, test ifadesi yanlış olana kadar tekrar eder.

For döngüsü yineleme sayısı geçerliği olduğu sürece kullanılır.  

For döngüsü, temel olarak bir kod bloğunu belirli bir sayıda ve üst üste çalıştırmak için kulllanılan bir döngüdür.

````c++
#include <iostream>
using namespace std;

for (int i = 1; i <= 10; i++) {
    cout << i << " ";
}
````

````c++
#include <iostream>
using namespace std;

int index;
cin >> index;

for (int i = 0; i <= index; i++) {
    if (i % 2 == 0) {
        cout << i << " ";
    }
}
````

### Döngülerden Çıkmak : Break ve Continue
* **Break :** Döngüyü sonlandırır. (Döngü dışına çıkılır.)
* **Continue :** Continue keyword'ünün altındaki kod kısımları atlanıp döngü devam ettirilir.

````c++
#include <iostream>
using namespace std;

for (int i = 0; i <= 5; i++)
{
    if (i == 3)
    {
        break;
    }
    cout << i << " ";
}
````

````c++
#include <iostream>
using namespace std;

for (int i = 0; i <= 5; i++)
{
    if (i == 3)
    {
        continue;
    }
    cout << i << " ";
}
````
